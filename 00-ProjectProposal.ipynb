## Research Question

How influential are the following factors: public buses and metros, percentage of people smoking, EV car usage, and number of factories in the city and suburbs, contributing toward air quality and the rate of reported asthma attacks in the population of asthma patients in major cities such as Tokyo, London, New York, Beijing and Paris? 

## Background and Prior Work

Transportation is a major contributor to air pollution worldwide, particularly in urban areas where vehicle density and traffic congestion are highest. Emissions from cars, trucks, and other private vehicles like particulate matter, PMO, nitrogen dioxide, NO2,  carbon monoxide, CO, etc. have been linked to lung disease from the decrease in air quality they cause. Cars are one of the largest contributors to poor air quality, making transportation a critical focus of air pollution mitigation. Public transportation has been a counter to private transportation like cars for decades and we want to investigate the relationship between public transportation and air quality in comparison to private transportation’s footprint and we want to use air quality indices and lung disease rates to make our analysis of the relationship.

Reference 1: https://www.uitp.org/news/why-public-transport-is-vital-in-achieving-clean-air-for-blue-skies/
	One relevant prior work is the article “Why Public Transport is Vital in Achieving Clean Air for Blue Skies”, published by the International Association of Public Transport (UITP). This article highlights how expanding public transportation reduces urban air pollution, improving public health. The article draws from the WHO and the United Nations Statistics Division to show that 99% of people breathe unsafe air, contributing to premature deaths. This article relates closely to our proposed project because it explores a connection between air quality and health, as well as the impact of public transportation on both. 

Reference 2: https://rosap.ntl.bts.gov/view/dot/62326
Another relevant prior work is the technical report "Assessing the Impact of Air Pollution on Public Health along Transit Routes”, which examines how transportation and socioeconomic factors influence respiratory health in some major metropolitan areas, Dallas-Fort Worth and LA. The researchers applied stat techniques such as PCA and OLS to identify predictors of respiratory disease. Their findings showed strong correlations between both demographics and access to transit/automobile access to respiratory disease risk. Interestingly, populations with higher transit access were more prone to respiratory diseases. This work relates closely to our proposed project because it demonstrates a clear connection between transportation systems, air quality, and respiratory health. 


## Hypothesis

When comparing the factors and how influential each can be, we hypothesized that the number of factories can be the most influential on how the air quality is, and asthma attacks will also increase as more people living in the area and working in the area will be affected.

 - [X] **A.2 Collection bias**: Have we considered sources of bias that could be introduced during data collection and survey design and taken steps to mitigate those?

Cultural and regulatory differences (such as smoking bans or EV incentives) further bias how data is collected and reported across cities.
