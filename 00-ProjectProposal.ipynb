## Background and Prior Work

Transportation is a major contributor to air pollution worldwide, particularly in urban areas where vehicle density and traffic congestion are highest. Emissions from cars, trucks, and other private vehicles like particulate matter, PMO, nitrogen dioxide, NO2,  carbon monoxide, CO, etc. have been linked to lung disease from the decrease in air quality they cause. Cars are one of the largest contributors to poor air quality, making transportation a critical focus of air pollution mitigation. Public transportation has been a counter to private transportation like cars for decades and we want to investigate the relationship between public transportation and air quality in comparison to private transportation’s footprint and we want to use air quality indices and lung disease rates to make our analysis of the relationship.

### A. Data Collection
 - [X] **A.1 Informed consent**: If there are human subjects, have they given informed consent, where subjects affirmatively opt-in and have a clear understanding of the data uses to which they consent?

> This project relies on aggregated, city-level secondary data (e.g., transportation usage, smoking rates, EV adoption, air quality metrics) rather than individual-level human subject data. Because we are not directly collecting data from individuals, informed consent is not applicable in the traditional sense. However, we acknowledge that the original data sources may have their own consent limitations, and we rely only on publicly available or ethically collected datasets.

 - [X] **A.2 Collection bias**: Have we considered sources of bias that could be introduced during data collection and survey design and taken steps to mitigate those?

> There are several sources of potential collection bias in this dataset. Large cities differ significantly in transportation budgets, population density, and infrastructure, which affects public transportation usage, EV adoption, and driving behavior. Denser cities may naturally rely less on private vehicles, while smoking rates may be influenced by density-related stressors or local social norms. Income-related bias is also present, as wealthier populations are more likely to own EVs and rely less on public transportation, while lower-income groups may experience higher stressors, higher smoking rates, and greater exposure to pollution. Cultural and regulatory differences (such as smoking bans or EV incentives) further bias how data is collected and reported across cities.

 - [ ] **A.3 Limit PII exposure**: Have we considered ways to minimize exposure of personally identifiable information (PII) for example through anonymization or not collecting information that isn't relevant for analysis?
 - [X] **A.4 Downstream bias mitigation**: Have we considered ways to enable testing downstream results for biased outcomes (e.g., collecting data on protected group status like race or gender)?

> While individual protected attributes such as race or gender are not directly included, we acknowledge that variables like income, age, transportation access, and neighborhood-level pollution can act as proxies for protected groups.
